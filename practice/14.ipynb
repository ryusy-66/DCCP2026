```python
"""텍스트 파일 단어 빈도"""
# 문제 : 영문 텍스트를 읽어 단어 빈도를 세고, 상위 5개를 출력하시오.

freq = {}
with open("speech.txt", "r", encoding="utf-8") as f:
    for line in f:
        for word in line.lower().split():
            word = word.strip(".,!?")
            if word:
                freq[word] = freq.get(word, 0) + 1

#빈도순 정렬
ranked = sorted(freq.items(), key=lambda x: x[1], reverse=True)
for word, count in ranked[:5]:
    print(f"{word}: {count}")
    
```

```python
"""CSV -> JSON 변환"""
#문제 : CSV로 저장된 학생 명부를 JSON으로 변환해 저장하시오.

import csv, json

#1) CSV 읽기 --- 딕셔너리 리스트로
students = []
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        students.append(row)

#2) JSON으로 저장
with open("students.json", "w", encoding="utf-8") as f:
    json.dump(students, f, ensure_ascii=False, indent=2)

    
```

```python
"""시 파일 행 단위 분석"""

import json

stats = []
with open ("poem.txt", "r", encoding="utf-8") as f:
    for i, line in enumerate(f, start=1):
        line = line.rstrip()
        stats.append({
            "행번호": i "내용": line, 
            "글자수": len(line), "어절수": len(line.split()), 
        })

with open ("poem_stats.json", "w", encoding="utf-8") as f:
    json.dump(stats, f, ensure_ascii=False, indent=2)

```

```python
"""안전한 정수 변환 함수"""
#문제 : 문자열을 정수로 변환하되, 실패하면 None을 돌려주시오.

def safe_int(s:str) -> int | None:
    try:
        return int(s)
    except (ValueError, TypeError):
        return None

print(safe_int("42"))
print(safe_int("hello"))
print(safe_int(None))
```

```python
"""사전 검색 (KeyError 처리)"""

dictionary = {"사랑": "강한 애정과 관심", "우정": "서로 돕고 아끼는 관계"}

def lookup(word:str) -> str:
    try:
        return dictionary[word]
    except KeyError:
        return f"'{word}'은(는) 사전에 없습니다."
    
print(lookup("사랑"))
print(lookup("희망"))
```

```python
"""여러 파일 일괄 처리"""

import logging
logging.basicConfig(level=logging.INFO, format = "[%(levelname)s] %(message)s")

paths = ["a.txt", "b.txt", "c.txt", "d.txt"]
total_chars = 0
for path in paths:
    try:
        with open(path, "r", encoding="utf-8") as f:
            text = f.read()
        total_chars += len(text)
        logging.info(f"{path}: {len(text)}자")
    except FileNotFoundError:
        logging.warning(f"{path}: 파일이 없어 건너뜁니다")
    except UnicodeDecodeError:
        logging.error(f"{path}: 인코딩 오류")

logging.info(f"합계: {total_chars}자")
```

```python
"""입력 검증 함수"""

def parse_score(text:str) -> int:
    """점수 문자열을 0~100 정수로 변환"""
    try:
        score = int(text)
    except ValueError:
        raise ValueError(f"숫자가 아닙니다: {text!r}")
    if not 0 <= score <= 100:
        raise ValueError (f"0~100 범위가 아닙니다: {score}")
    return score

#사용 예
for raw in ["85", "120", "A+", "0"]:
    try:
        print (parse_score(raw))
    except ValueError as e:
        print (f"오류: {e}")
```

```python
"""디버거로 버그 찾기"""
#문제 : 다음 함수는 평균을 잘못 계산한다. breakpoint()로 원인을 찾으시오.

def buggy_average(nums: list[int]) -> float:
    total = 0
    for i, n in enumerate(nums):
        if n%2 == 0:
            total += n
    breakpoint()
    return total/len(nums)

print (buggy_average([10, 20, 30, 40]))
```
